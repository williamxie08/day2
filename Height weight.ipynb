```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

filename = 'SOCR-HeightWeight.csv'
df = pd.read_csv(filename, index_col=0)
print(df)

height = np.array(df['Height(Inches)'])
weight = np.array(df['Weight(Pounds)'])

CM_IN_RATIO = 2.54
KG_LBS_RATIO = 0.453592

height_cm = height * CM_IN_RATIO
weight_kg = weight * KG_LBS_RATIO

tallest = np.max(height_cm)
shortest = np.min(height_cm)
heaviest = np.max(weight_kg)
lightest = np.min(weight_kg)

plt.figure(figsize=(10,6))
plt.scatter(height_cm, weight_kg)
plt.xlabel('height(CM)')
plt.ylabel('Weight(KG)')
plt.grid(True)
plt.show()

height = height_cm
weight = weight_kg

def predict_weight(height, a, b):
    weight = a * height + b
    return weight

weight_pred = predict_weight(height, 0.9, -100)

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

def mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

print(mse(weight, weight_pred))
print(mae(weight, weight_pred))

```
